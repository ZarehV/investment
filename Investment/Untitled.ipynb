```python
import yfinance as yf
from common.strategies import ScalpingVWAPRSI
```

```python
ticker = 'IJR'
data = yf.download(ticker, period='1y', interval='1d')
```

```python
strategy = ScalpingVWAPRSI(data, 'Adj Close', False)  
strategy.get_indicators()
strategy.get_signals(
    backcandles=15,     
    use_bb=False,
    use_rsi=False
)
strategy.backtest()
```
