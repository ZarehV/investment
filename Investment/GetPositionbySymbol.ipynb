# Get Position by Symbol

```python
symbol = input("Povide symbol?")
from common.journal import get_journal

journal = get_journal()
journal_new = journal[journal["symbol"] == symbol]
if len(journal_new.index) > 0:
    for index, row in journal_new.iterrows():
        print("*" * 10)
        print(row["symbol"])
        print(row["investment_account"])
        print(row["status"])
        print(row["key"])
        print("*" * 10)

else:
    print(f"No enties for symbol {symbol}")
```
