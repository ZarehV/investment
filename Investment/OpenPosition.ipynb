```python
from datetime import datetime
from common.routines import prep_position
from common.calculations import calculate_porfit_loss, calculate_recommended_take_profit, calculate_porfit_loss, calculate_risk_reward_ratio
from common.journal import printrecord, save_new_entry
from common.utils import compute_hash
```

```python
position_size = 266
ticker = 'VYM'
entry_price, stop_loss, take_profit, risk_reward_ratio = prep_position(
    ticker=ticker, 
    days=10,
    investment_type='Long Term Investment',
    #investment_type='afternoon',
    testing=False,
    atr_window=4,
    ema_period=20,
    print_metrics=False,
    #risk_level="low",
    risk_level="medium",
    position_type='long',
    stop_loss_type='ema',
    take_profit_type='resistance',
    #plot=False,
    plot=False,
    debug=False
)
```

```python
taget_profit_dls, taget_profit_pct = calculate_porfit_loss(entry_price, take_profit, position_size)
taget_loss_dls, taget_loss_pct = calculate_porfit_loss(entry_price, stop_loss, position_size)
new_investment = {
    "symbol": ticker,
    "purchase_price": entry_price,
    "position_size": position_size,
    "purchase_date": datetime.now().strftime("%Y-%m-%d"),
    "investment_account": "",
    "investment_reason": "",
    "investment_type": "Long",
    "stop_loss": stop_loss,
    "target_take_profit": take_profit,
    "target_profit": taget_profit_dls,
    "target_profit_pct": taget_profit_pct,
    "target_loss": taget_loss_dls,
    "target_loss_pct": taget_loss_pct,    
    "target_risk_reward_ratio": risk_reward_ratio,     
    "status": "open",
    "actual_profit_lost": 0.0,
    "actual_profit_lost_pct": 0.0,
    "exit_comments": " "
}
```

```python
printrecord(new_investment)
```

```python
new_investment['purchase_price'] = 127.68
new_investment['stop_loss'] = 122
new_investment['target_take_profit'] = calculate_recommended_take_profit(ticker, new_investment['purchase_price'], new_investment['stop_loss'], risk_reward_ratio=2)
new_investment['target_profit'], new_investment['target_profit_pct'] = calculate_porfit_loss(
    new_investment['purchase_price'], 
    new_investment['target_take_profit'], 
    position_size
)
new_investment['target_loss'], new_investment['target_loss_pct'] = calculate_porfit_loss(
    new_investment['purchase_price'], 
    new_investment['stop_loss'], 
    position_size
)
new_investment['target_risk_reward_ratio'] = calculate_risk_reward_ratio(
    new_investment['purchase_price'], 
    new_investment['stop_loss'], 
    new_investment['target_take_profit']
)
new_investment['key'] = compute_hash(new_investment)
printrecord(new_investment)
```

```python
save_new_entry(new_investment)
```
